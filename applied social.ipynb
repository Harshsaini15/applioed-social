```python
import networkx as nx
import matplotlib.pyplot as plt

# Create graph
G = nx.Graph()
G.add_edges_from([(1,2), (2,3), (3,4), (4,5), (5,1), (2,4)])

# Calculate assortativity
assortativity = nx.degree_assortativity_coefficient(G)
print("Assortativity Coefficient:", assortativity)

# Node colors based on degree
degrees = dict(G.degree())
node_colors = [degrees[n] for n in G.nodes()]

# Draw graph
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=2000, cmap=plt.cm.viridis)
plt.title("Graph with Degree-based Coloring")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create DAG
G = nx.DiGraph()
G.add_edges_from([
    ("A", "C"),
    ("B", "C"),
    ("C", "D"),
    ("D", "E")
])

# Check DAG
print("Is DAG?", nx.is_directed_acyclic_graph(G))

# Topological sort
topo_order = list(nx.topological_sort(G))
print("Topological Order:", topo_order)

# Draw graph (layered look)
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, arrows=True)

# Show order on graph
labels = {node: f"{node}\n({i})" for i, node in enumerate(topo_order)}
nx.draw_networkx_labels(G, pos, labels)

plt.title("DAG with Topological Order")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create flow network
G = nx.DiGraph()

# Add edges with capacities
G.add_edge('S', 'A', capacity=10)
G.add_edge('S', 'C', capacity=10)
G.add_edge('A', 'B', capacity=4)
G.add_edge('A', 'C', capacity=2)
G.add_edge('C', 'D', capacity=9)
G.add_edge('B', 'T', capacity=10)
G.add_edge('D', 'B', capacity=6)
G.add_edge('D', 'T', capacity=10)

# Compute max flow
flow_value, flow_dict = nx.maximum_flow(G, 'S', 'T')

print("Maximum Flow Value:", flow_value)
print("Flow Distribution:", flow_dict)

# Draw graph
pos = nx.spring_layout(G, seed=42)

nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightgreen', arrows=True)

# Show capacities
edge_labels = nx.get_edge_attributes(G, 'capacity')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Flow Network (Capacities on Edges)")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

# Create signed graph
G = nx.Graph()

# Add edges with sign (+1 or -1)
edges = [
    (1, 2, 1),
    (2, 3, -1),
    (3, 1, 1),
    (3, 4, -1),
    (4, 1, 1)
]

G.add_weighted_edges_from(edges)

# Function to find unstable triangles
def find_unstable_triangles(G):
    unstable = []
    for triangle in combinations(G.nodes(), 3):
        if G.has_edge(triangle[0], triangle[1]) and \
           G.has_edge(triangle[1], triangle[2]) and \
           G.has_edge(triangle[2], triangle[0]):
            
            sign_product = (
                G[triangle[0]][triangle[1]]['weight'] *
                G[triangle[1]][triangle[2]]['weight'] *
                G[triangle[2]][triangle[0]]['weight']
            )
            
            if sign_product == -1:
                unstable.append(triangle)
    
    return unstable

unstable_triangles = find_unstable_triangles(G)
print("Unstable Triangles:", unstable_triangles)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

# Create signed graph
G = nx.Graph()

edges = [
    (1, 2, 1),
    (2, 3, -1),
    (3, 1, 1),
    (3, 4, -1),
    (4, 1, 1)
]

G.add_weighted_edges_from(edges)

# Function to find unstable triangles
def find_unstable_triangles(G):
    unstable = []
    for tri in combinations(G.nodes(), 3):
        if G.has_edge(tri[0], tri[1]) and \
           G.has_edge(tri[1], tri[2]) and \
           G.has_edge(tri[2], tri[0]):
            
            prod = (
                G[tri[0]][tri[1]]['weight'] *
                G[tri[1]][tri[2]]['weight'] *
                G[tri[2]][tri[0]]['weight']
            )
            
            if prod == -1:
                unstable.append(tri)
    return unstable

unstable_triangles = find_unstable_triangles(G)
print("Unstable Triangles:", unstable_triangles)

# --- DRAW GRAPH ---
pos = nx.spring_layout(G, seed=42)

edge_colors = ['green' if G[u][v]['weight']==1 else 'red' for u,v in G.edges()]

nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue')
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=2)

# Highlight unstable triangles
for tri in unstable_triangles:
    nx.draw_networkx_nodes(G, pos, nodelist=tri, node_color='yellow')

nx.draw_networkx_edge_labels(G, pos,
    edge_labels={(u,v):G[u][v]['weight'] for u,v in G.edges()})

plt.title("Unstable Triangles Highlighted (Yellow Nodes)")
plt.show()
```

```python
# Count unstable triangles
count = len(unstable_triangles)
print("Number of Unstable Triangles:", count)

# Draw again with title showing count
plt.figure()

nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue')
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=2)

plt.title(f"Graph with {count} Unstable Triangle(s)")
plt.show()
```

```python
# Function to stabilize network
def make_network_stable(G):
    G = G.copy()
    unstable = find_unstable_triangles(G)
    
    for tri in unstable:
        u, v, w = tri
        # Flip one edge
        G[u][v]['weight'] *= -1
    
    return G

# BEFORE graph
plt.figure()
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue')
nx.draw_networkx_edges(G, pos, edge_color=edge_colors, width=2)
plt.title("Before Stabilization")
plt.show()

# Apply fix
G_stable = make_network_stable(G)

# AFTER graph
pos2 = nx.spring_layout(G_stable, seed=42)

edge_colors2 = ['green' if G_stable[u][v]['weight']==1 else 'red' for u,v in G_stable.edges()]

plt.figure()
nx.draw(G_stable, pos2, with_labels=True, node_size=2000, node_color='lightgreen')
nx.draw_networkx_edges(G_stable, pos2, edge_color=edge_colors2, width=2)

nx.draw_networkx_edge_labels(G_stable, pos2,
    edge_labels={(u,v):G_stable[u][v]['weight'] for u,v in G_stable.edges()})

plt.title("After Stabilization (No Unstable Triangles)")
plt.show()

# Verify
print("Unstable after fixing:", find_unstable_triangles(G_stable))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import community

# Create graph
G = nx.karate_club_graph()  # famous dataset

# Detect communities
communities = community.greedy_modularity_communities(G)

# Convert to list
communities = [list(c) for c in communities]
print("Communities:", communities)

# Assign colors
color_map = []
for node in G:
    for i, comm in enumerate(communities):
        if node in comm:
            color_map.append(i)

# Draw graph
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, node_color=color_map, with_labels=True, cmap=plt.cm.Set3, node_size=500)

plt.title("Community Detection in Graph")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

# Create graph
G = nx.Graph()
G.add_edges_from([(1,2), (2,3), (3,4)])

# BEFORE
pos = nx.spring_layout(G, seed=42)
plt.figure()
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue')
plt.title("Before Triadic Closure")
plt.show()

# Apply triadic closure
new_edges = []
for u, v, w in combinations(G.nodes(), 3):
    if G.has_edge(u,v) and G.has_edge(v,w) and not G.has_edge(u,w):
        new_edges.append((u,w))

G.add_edges_from(new_edges)

print("Edges added:", new_edges)

# AFTER
plt.figure()
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightgreen')
plt.title("After Triadic Closure")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create weighted graph
G = nx.Graph()

edges = [
    (1,2,5), (2,3,4), (3,1,5),   # strong cluster
    (3,4,1),                     # weak bridge
    (4,5,5), (5,6,4), (6,4,5)    # another cluster
]

G.add_weighted_edges_from(edges)

# Define threshold
threshold = 3

strong_edges = [(u,v) for u,v,d in G.edges(data=True) if d['weight'] >= threshold]
weak_edges = [(u,v) for u,v,d in G.edges(data=True) if d['weight'] < threshold]

pos = nx.spring_layout(G, seed=42)

plt.figure()

# Draw nodes
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='lightblue')

# Draw strong ties (thick)
nx.draw_networkx_edges(G, pos, edgelist=strong_edges, width=3, edge_color='green')

# Draw weak ties (thin dashed)
nx.draw_networkx_edges(G, pos, edgelist=weak_edges, width=1, edge_color='red', style='dashed')

nx.draw_networkx_labels(G, pos)

# Edge labels
labels = {(u,v):d['weight'] for u,v,d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.title("Strong (Green) vs Weak (Red Dashed) Ties")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create graph
G = nx.Graph()
G.add_edges_from([(1,2), (1,3), (2,3), (3,4), (4,5)])

# Initialize points
points = {node: 1 for node in G.nodes()}

# Perform iterations
for _ in range(5):
    new_points = {node: 0 for node in G.nodes()}
    
    for node in G.nodes():
        neighbors = list(G.neighbors(node))
        if neighbors:
            share = points[node] / len(neighbors)
            for nbr in neighbors:
                new_points[nbr] += share
    
    points = new_points

print("Final Points Distribution:", points)

# Visualization
pos = nx.spring_layout(G, seed=42)
sizes = [points[n]*3000 for n in G.nodes()]

nx.draw(G, pos, with_labels=True, node_size=sizes, node_color='lightblue')
plt.title("Points Distribution (Node Importance)")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Create graph
G = nx.Graph()
G.add_edges_from([(1,2), (2,3), (3,4), (4,5), (3,5), (2,4)])

# Random walk
start = 1
walk = [start]

for _ in range(10):
    current = walk[-1]
    neighbors = list(G.neighbors(current))
    next_node = random.choice(neighbors)
    walk.append(next_node)

print("Random Walk Path:", walk)

# Visualization
pos = nx.spring_layout(G, seed=42)

plt.figure()
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue')

# Highlight path
path_edges = list(zip(walk, walk[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='red', width=3)

plt.title("Random Walk Path Highlighted")
plt.show()
```
